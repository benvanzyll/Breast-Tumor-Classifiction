```python
# Understand how certain features may affect the outcome

breastCancer.groupby('diagnosis').mean()
```

```python
# Create features dataframe

features = breastCancer.drop('diagnosis', axis=1)
features.head()
```

```python
# Encode binary outcome (0 for benign, 1 for malignant)

diagnosis_num = {'B':0, 'M':1}
breastCancer['diagnosis'] = breastCancer['diagnosis'].map(diagnosis_num)
breastCancer.head()
```

```python
# Create labels dataframe

labels = breastCancer['diagnosis']
labels.head()
```

```python
# Create correlation matrix for features with 'mean' attributes

correlation_means = features[['radius_mean', 'texture_mean', 'perimeter_mean',
                      'area_mean', 'smoothness_mean', 'compactness_mean',
                      'concavity_mean', 'concave points_mean', 'symmetry_mean',
                      'fractal_dimension_mean']].corr()
sns.heatmap(correlation_means, xticklabels = correlation_means.columns, 
            yticklabels = correlation_means.columns, annot = True)
```

```python
# Create correlation matrix for features with 'worst' attributes

correlation_worst = features[['radius_worst', 'texture_worst', 'perimeter_worst',
                      'area_worst', 'smoothness_worst', 'compactness_worst',
                      'concavity_worst', 'concave points_worst', 'symmetry_worst',
                      'fractal_dimension_worst']].corr()
sns.heatmap(correlation_worst, xticklabels = correlation_worst.columns, 
            yticklabels = correlation_worst.columns, annot = True)
```

```python
# Create pair plots for features with 'mean' attributes

sns.pairplot(features[['radius_mean', 'texture_mean', 'perimeter_mean',
                      'area_mean', 'smoothness_mean', 'compactness_mean',
                      'concavity_mean', 'concave points_mean', 'symmetry_mean',
                      'fractal_dimension_mean']])
```

```python
# Create pair plots for features with 'worst' attributes

sns.pairplot(features[['radius_worst', 'texture_worst', 'perimeter_worst',
                      'area_worst', 'smoothness_worst', 'compactness_worst',
                      'concavity_worst', 'concave points_worst', 'symmetry_worst',
                      'fractal_dimension_worst']])
```

```python
# Create a scatterplot showing relationship between radius mean and perimeter mean

sns.relplot(x='radius_mean', y='perimeter_mean', hue='diagnosis', data=breastCancer)
```
